```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.flipkart.com/dotd-store?=Web&wid=1.dealCard.OMU_1&otracker=clp_omu_Deals%2Bof%2Bthe%2BDay_offers-store_1&otracker1=clp_omu_PINNED_neo%2Fmerchandising_Deals%2Bof%2Bthe%2BDay_NA_wc_view-all_1')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name = []
for i in soup.find_all('div',class_="_3LU4EM"):
    name.append(i.text)

name
```

```python
discount = []
for i in soup.find_all('div',class_="_2tDhp2"):
    discount.append(i.text)
print(discount, len(discount))
```

```python
image = []
for i in soup.find_all('img',class_="_396cs4 _3exPp9"):
    image.append(i['src'])
print(image,len(image))
```

```python
import pandas as pd
df = pd.DataFrame({"Product Name":name,"Offer":discount,"Photo":image})
df
```
