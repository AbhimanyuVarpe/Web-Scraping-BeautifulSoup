```python
# write in location bar
search_location.send_keys('Bangalore')
```

**3. Then click the search button.**

```python
# clicking search button by using xpath
search_bttn = driver.find_element_by_xpath("/html/body/div[1]/div[3]/div[2]/section/div/form/div[3]/button")
search_bttn
```

```python
search_bttn.click()
```

**4. Then scrape the data for the first 10 jobs results you get.**

```python
# Extracting job title

job_title = [] # creating empty list

title = driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')[:10]
print(len(title))

for i in title:
    job_title.append(i.text)
    
len(job_title)
```

```python
# Extracting job_location

job_location = [] # creating empty list

loc = driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')[:10]
print(len(loc))

for i in loc:
    job_location.append(i.text)
    
len(job_location)
```

```python
# Extracting company_name

company_name = [] # creating empty list

comp = driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')[:10]
print(len(comp))

for i in comp:
    company_name.append(i.text)
    
len(company_name)
```

```python
# Extracting experience required

experience = [] # creating empty list

exp = driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')[:10]
print(len(exp))

for i in exp:
    experience.append(i.text)
    
len(experience)
```

**5. Finally create a dataframe of the scraped data.**

```python
df = pd.DataFrame({'Job Title': job_title,'Company Name':company_name,'Job location':job_location,'Experience required':experience})
df
```

### Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link: https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace.

**As shown in the above page you have to scrape the tick marked attributes.These are:**

4. You have to **scrape this data for first 100 reviews.**

```python
# connecting to the driver
driver = webdriver.Chrome('chromedriver.exe')

# getting to webpage
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace')
```

```python
# clicking on all reviews using xpath
clk_all = driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div/div/div[5]/div/a/div')

clk_all.click()
```

```python
# Extracting data

#creating empty lists
Rating = []
Summary = []
Full_review = []

#scrapping the required details

for page in range(10): # loop for scrapping 3 page
        
    url = driver.find_element_by_xpath('//a[@class="_1LKTO3"]').get_attribute('href')
    driver.get(url)
    
    rating = driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]') #scraping Ratings by class name = '_3LWZlK _1BLPMq'
    for i in rating:
        Rating.append(i.text)
        
    summary = driver.find_elements_by_class_name('_2-N8zT') #scraping Review summary by class name = '_2-N8zT'
    for j in summary:
        Summary.append(j.text)
    
    full = driver.find_elements_by_class_name('t-ZTKy') #scraping Review summary by class name = 't-ZTKy'
    for k in full:
        Full_review.append(k.text)

    
print(len(Rating),
len(Summary),
len(Full_review))
```

```python
#Combining all the lists into a single dataframe
df=pd.DataFrame({'Rating': Rating ,
                'Review Summary': Summary,
               'Full review': Full_review})
df
```

### Q7: Go to the link - https://www.myntra.com/shoes

And then scrape First 100 shoes data you get. The data should include “Brand” of the shoes , Short Shoe description, price of the shoe as shown in the below image.

```python
# connecting to the web driver
driver = webdriver.Chrome('chromedriver.exe')

# Getting to the web page
driver.get('https://www.myntra.com/shoes')
```

```python
# Appling the filter for Color “Black”
color_filter = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
color_filter.click()
```

```python
# Appling the filter for price “Rs. 7149 to Rs. 14099”
price_filter = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
price_filter.click()
```

```python
# Creating empty lists
Brand = []
Description = []
Price = []
```

```python
# scraping the required details

#Taking 2 pages into consideration using for loop
for page in range(2):
    url = driver.find_element_by_xpath('//a[@rel="next"]').get_attribute('href') 
    driver.get(url)
    
    # scraping brand name by class name
    brand = driver.find_elements_by_xpath('//h3[@class="product-brand"]')
    for i in brand:
        Brand.append(i.text)
    
    # scraping product description by class name
    desc = driver.find_elements_by_xpath('//h4[@class="product-product"]')
    for i in desc:
        Description.append(i.text)
    
    # scraping price by class name
    price = driver.find_elements_by_xpath('//div[@class="product-price"]')
    for i in price:
        Price.append(i.text.replace("Rs.","").split(" ")[1])
        
print(len(Brand),len(Description),len(Price))
```

```python
df = pd.DataFrame({'Brand': Brand ,'Product Description':Description , 'Price in Rs.': Price })
df
```

```python
# scrapping required data

name = [] # creating empty lists
data = []
min_sal = []
max_sal = []
avg_sal = []
exp = []

# Scraping name of company and Quantity of data
N = driver.find_elements_by_xpath('//div[@class="name"]')[:10]
for i in N:
    name.append(i.text.split("\n")[0])
    data.append(i.text.split("\n")[1])
    
# Scraping Experience
E = driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')[:10]
for j in E:
    exp.append(j.text.split("\n")[-1])

# Scraping Salary data
minimum = driver.find_elements_by_xpath('//div[@class="value body-medium"]')[0::2]
for k in minimum:
    min_sal.append(k.text)
    
maximum = driver.find_elements_by_xpath('//div[@class="value body-medium"]')[1::2]
for k in maximum:
    max_sal.append(k.text)
    
average = driver.find_elements_by_xpath('//p[@class="averageCtc"]')[:10]
for k in average:
    avg_sal.append(k.text)
    
print(len(name),
len(data),
len(min_sal),
len(max_sal),
len(avg_sal),
len(exp))
```
